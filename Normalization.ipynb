```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
link="https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day25-normalization/wine_data.csv"
df=pd.read_csv(link,header=None,usecols=[0,1,2])
df.columns=["class_label","alcohol","malic_acid"]
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.kdeplot(df["alcohol"])
plt.title("Alcohol")
plt.subplot(1,2,2)
sns.kdeplot(df["malic_acid"])
plt.title("malic_acid")
```

```python
x=df.iloc[:,1:]
y=df["class_label"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
x_train.shape,y_train.shape
```

```python
mms=MinMaxScaler()
x_train_scaled=mms.fit_transform(x_train)
x_test_scaled=mms.transform(x_test)
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_train_scaled
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.scatterplot(x=x_train["alcohol"], y=x_train["malic_acid"],data=x_train,hue=y_train,palette="dark")
plt.title("Before")
plt.subplot(1,2,2)
sns.scatterplot(x=x_train_scaled["alcohol"],y=x_train["malic_acid"],data=x_train_scaled,hue=y_train,palette="dark")
plt.title("After")
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.kdeplot(x_train["alcohol"])
plt.title("Before")
plt.subplot(1,2,2)
sns.kdeplot(x_train_scaled["alcohol"])
plt.title("After")
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.kdeplot(x_train["malic_acid"])
plt.title("Before")
plt.subplot(1,2,2)
sns.kdeplot(x_train_scaled["malic_acid"])
plt.title("After")
```
