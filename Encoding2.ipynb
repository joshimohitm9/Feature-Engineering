```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.town=le.fit_transform(df.town)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
town=ohe.fit_transform(df[["town"]]).toarray()
town
```

```python
town=pd.DataFrame(town)
town
```

```python
df2=df[["area","price"]]
df2
```

```python
df=pd.concat([town,df2],axis=1)
```

```python
df.columns=["monroe township","west windsor","robinsville","area","price"]
```

```python
df
```

```python
df.drop("robinsville",axis="columns",inplace=True)
```

```python
x=df.iloc[:,0:3]
x
```

```python
y=df["price"]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=56)
```
